### My code plots a bar graph with the amount of dogs, classified by their breeds in a neighborhood (random data)

```python
import random
import matplotlib.pyplot as plt
x = ["Bulldog", "Boxer", "Pitbull", "Poodle", "Husky", "Beagle", "Labrador"]
y = [random.randint(1,20) for i in range(7)]
plt.bar(x, y)
plt.xlabel("Dog Breeds", fontweight = "bold")
plt.ylabel("Quantity", fontweight = "bold")
plt.title("Quantity of dogs by breed", fontweight = "bold")
plt.show()
```
